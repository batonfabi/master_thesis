```python
import IPython
import librosa
import soundfile
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
%matplotlib inline
```

## Beispiel der Berechnung und Anzeige eines Spektrogramms. Mit dem HÃ¶rbeispiel.

```python
# Laden der Audiodatei
wav_array,_ = soundfile.read('../audiobeispiele/kohlmeise.flac')

# Berechnung des Spektrogramms
spec = librosa.stft(wav_array)

# Anzeige des Spektrogramms
fig, ax = plt.subplots()
D = librosa.amplitude_to_db(np.abs(librosa.stft(wav_array)),ref=np.max)
img = librosa.display.specshow(D, y_axis='linear', x_axis='time',sr=22050, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
```
